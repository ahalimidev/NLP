```python
import pandas as pd

data = pd.read_excel('hasil/klasifikasi_sentimen_perasaan.xlsx')
data
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer


# Tokenize and preprocess the documents
documents = data['text_join']

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer()

# Compute TF-IDF
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

# Get feature names
feature_names = tfidf_vectorizer.get_feature_names_out()

# Create a dataframe of the TF-IDF representation
df_tfidf = pd.DataFrame(tfidf_matrix.toarray(), columns=feature_names)

# Save the dataframe to an Excel file
df_tfidf.to_excel('hasil/tf-idf.xlsx', index=False)

# Display the dataframe
print(df_tfidf)
```

# N-Grams

```python
import nltk
from nltk.util import ngrams
from collections import Counter

# Tokenize and preprocess the documents
documents = data['text_join']

# Function to generate N-Grams
def generate_ngrams(tokens, n):
    return list(ngrams(tokens, n))

# Preprocess and tokenize the documents
tokenized_documents = [nltk.word_tokenize(doc.lower()) for doc in documents]

# Generate and count unigrams, bigrams, and trigrams
unigrams = [gram for doc in tokenized_documents for gram in generate_ngrams(doc, 1)]
bigrams = [gram for doc in tokenized_documents for gram in generate_ngrams(doc, 2)]
trigrams = [gram for doc in tokenized_documents for gram in generate_ngrams(doc, 3)]

# Count the occurrences of each n-gram
unigram_counts = Counter(unigrams)
bigram_counts = Counter(bigrams)
trigram_counts = Counter(trigrams)

# Convert counts to DataFrame
unigram_df = pd.DataFrame(unigram_counts.items(), columns=['Unigram', 'Count'])
unigram_df.to_excel('hasil/ungram.xlsx', index=False)

bigram_df = pd.DataFrame(bigram_counts.items(), columns=['Bigram', 'Count'])
bigram_df.to_excel('hasil/bigram.xlsx', index=False)

trigram_df = pd.DataFrame(trigram_counts.items(), columns=['Trigram', 'Count'])
bigram_df.to_excel('hasil/trigram.xlsx', index=False)


# Print DataFrames
print("Unigrams DataFrame:")
print(unigram_df.head())
print("\nBigrams DataFrame:")
print(bigram_df.head())
print("\nTrigrams DataFrame:")
print(trigram_df.head())
```

# BOW

```python
from sklearn.feature_extraction.text import CountVectorizer

# Contoh data sentimen
documents = data['text_join']


# Bag-of-Words
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(documents)

# Matriks representasi BOW
bow_matrix = X.toarray()

# Kata-kata unik dalam vektor fitur
feature_names = vectorizer.get_feature_names_out()

# Membuat DataFrame
df_bow = pd.DataFrame(data=bow_matrix, columns=feature_names)
df_bow.to_excel('hasil/bow.xlsx', index=False)

print("DataFrame representasi BOW:")
print(df_bow)
```
